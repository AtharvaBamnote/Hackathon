```python
import asyncio
from telegram import Update
from telegram.ext import Application, MessageHandler, CallbackContext, filters
import whisper

# Initialize Telegram bot application (Replace 'YOUR_BOT_TOKEN' with your actual bot token)
app = Application.builder().token("YOUR_BOT_TOKEN").build()

# Load the Whisper model
model = whisper.load_model("base")

# Speech-to-text function
def speech_to_text(audio_path):
    result = model.transcribe(audio_path)
    return result["text"]

# Asynchronous function to handle voice messages
async def handle_voice(update: Update, context: CallbackContext):
    file = await update.message.voice.get_file()

    # Download the voice message
    audio_path = "voice.ogg"
    await file.download_to_drive(audio_path)

    # Convert speech to text
    text = speech_to_text(audio_path)
    await update.message.reply_text(f"You said: {text}")

    # Generate a story based on the extracted text
    story_text = generate_story(text)  # Ensure this function exists
    await update.message.reply_text(story_text)

# Add handler to process voice messages
app.add_handler(MessageHandler(filters.VOICE, handle_voice))

# Function to run the bot safely in Jupyter Notebook or Colab
async def main():
    print("Bot is running... 🚀")
    await app.run_polling()

# Fix for "RuntimeError: This event loop is already running"
if __name__ == "__main__":
    try:
        loop = asyncio.get_running_loop()
    except RuntimeError:
        loop = asyncio.new_event_loop()
        asyncio.set_event_loop(loop)

    loop.create_task(main())  # Start bot as a background task
```
