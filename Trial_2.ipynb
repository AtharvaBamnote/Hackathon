```python
# Load AI Models
print("Loading models, please wait...")
openai.api_key = OPENAI_API_KEY
whisper_model = whisper.load_model("base")  # Whisper for speech-to-text
```

```python
# Check if GPU is available
device = "cuda" if torch.cuda.is_available() else "cpu"
pipeline = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5").to(device)
```

```python
#Function to generate a story using OpenAI GPT
def generate_story(prompt):
    try:
        response = openai.ChatCompletion.create(
            model="gpt-4",
            messages=[{"role": "user", "content": f"Tell a captivating story about {prompt}."}]
        )
        return response["choices"][0]["message"]["content"]
    except Exception as e:
        return f"Oops! Something went wrong: {e}"
```

```python
# Function to generate an image using Stable Diffusion
def generate_image(prompt):
    try:
        image = pipeline(prompt).images[0]
        image_path = "generated_story_image.png"
        image.save(image_path)
        return image_path
    except Exception as e:
        print(f"Image generation error: {e}")
        return None
```

```python
# Function to convert speech to text using Whisper
def speech_to_text(audio_path):
    try:
        result = whisper_model.transcribe(audio_path)
        return result["text"]
    except Exception as e:
        return f"Error processing audio: {e}"
```

```python
# Telegram Bot Handlers
async def start(update: Update, context: CallbackContext):
    await update.message.reply_text("Welcome to *Dastan Bot*! \n\nTell me a topic, and I'll weave an enchanting story just for you! ")

async def story_with_image(update: Update, context: CallbackContext):
    prompt = update.message.text
    story_text = generate_story(prompt)
    image_path = generate_image(prompt)

    await update.message.reply_text(f" Here's your story on *{prompt}*:\n\n{story_text}")

    if image_path:
        with open(image_path, "rb") as photo:
            await update.message.reply_photo(photo=photo, caption=" Here's a visual representation of your story!")

async def handle_voice(update: Update, context: CallbackContext):
    file = await update.message.voice.get_file()
    audio_path = "voice_message.ogg"
    await file.download_to_drive(audio_path)

    text = speech_to_text(audio_path)
    await update.message.reply_text(f"You said: {text}")

    story_text = generate_story(text)
    await update.message.reply_text(f" Based on what you said, here's your story:\n\n{story_text}")
```

```python
# Initialize Telegram Bot Application
app = Application.builder().token(TELEGRAM_BOT_TOKEN).build()
```

```python
# Add Handlers
app.add_handler(CommandHandler("start", start))
app.add_handler(MessageHandler(filters.TEXT & ~filters.COMMAND, story_with_image))
app.add_handler(MessageHandler(filters.VOICE, handle_voice))
```

```python
# Run the bot asynchronously
async def main():
    print("Dastan Bot is now live on Telegram!")
    await app.run_polling()
```

```python
# Run the bot inside Colab's event loop
if __name__ == "__main__":
    try:
        loop = asyncio.get_running_loop()
    except RuntimeError:
        loop = asyncio.new_event_loop()
        asyncio.set_event_loop(loop)

    loop.create_task(main())  # Run the bot in the background
```
